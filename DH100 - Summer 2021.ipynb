```python
#Loading the dataframe in .csv

all_speeches = pd.read_csv("/Users/julianatorres/Desktop/DH/dataset/speeches.csv", index_col=0)

#Here is the entire data, followed by a sample of the first five speeches. 

print(all_speeches)
all_speeches.head()
```

```python
# Before creating a "master wordcloud" showing the most used words across all speeches, I created this wordcloud for FDRs first speech. 

speech_1 = all_speeches.Speech[1]

wordcloud = WordCloud(max_words=50).generate(speech_1)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Now I will combine all speeches and make one wordcloud for all of them

speeches_combined = " ".join(word for word in all_speeches.Speech)
print ("There are {} words in the combination of all speeches.".format(len(speeches_combined)))

```

```python
# Wordcloud for the combined speeches, minus the words in the wordstop list

wordcloud = WordCloud(stopwords=stopwords, max_words=50).generate(speeches_combined)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
 
```
